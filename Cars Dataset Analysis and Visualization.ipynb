```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

cars = pd.read_csv('cars.csv')
```

```python
cars
```

```python
cars.dtypes
```

```python
cars.describe()counts = cars['make'].value_counts()
counts
```

```python
counts = cars['make'].value_counts()
counts
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
counts.plot.bar(ax = ax)
plt.xlabel = ('Maker')
plt.ylabel = ('Number of autos')
plt.title = ('Number of autos by makers')
```

```python
cars['fueltype'].value_counts()
```

```python
sns.factorplot('fueltype',data=cars,kind='count')
```

```python
sns.factorplot(x='fueltype', y='price', data=cars, kind='bar')
```

```python
var = cars.groupby('make').price.mean()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('Brand')
ax1.set_ylabel('Price')
ax1.set_title("Brand Vs Price")
var.plot(kind='bar')
```

```python
sns.set_style("whitegrid")
sns.jointplot('horsepower', 'price', data = cars, alpha = 0.3)
```

```python
sns.regplot(data = cars,
            x = 'price',
            y = 'horsepower',)
```

```python
sns.regplot(data = cars,
            x = 'enginesize',
            y = 'horsepower',)
```

```python
plt.figure(figsize = (30,20))
sns.set(style='whitegrid')

plt.subplot(3,1,1)

sns.barplot(data = cars, x = 'make', y = 'price', color = 'green')
                                                    
plt.subplot(3,1,2)
sns.barplot(data = cars, x = 'make', y = 'enginesize', color = 'blue')

plt.subplot(3,1,3)
sns.barplot(data = cars, x = 'make', y = 'horsepower', color = 'red')
```

```python
jaguar = cars[cars.make == 'jaguar']
mercedes = cars[cars.make == 'mercedes']
porsche = cars[cars.make == 'porsche']

plt.plot(jaguar.horsepower, jaguar.price)
plt.plot(mercedes.horsepower, mercedes.price)
plt.plot(porsche.horsepower, porsche.price)

plt.legend(['Jaguar', 'Mercedes', 'Porsche'])
plt.title("Horsepower x Price")
plt.show()
```
